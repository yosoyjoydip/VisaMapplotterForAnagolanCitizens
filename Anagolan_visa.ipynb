```python
!pip install pandas pycountry plotly
import pandas as pd
import pycountry
import plotly.express as px
```

File named "passport-index-tidy-iso3.csv" is enough to fetch the required data

```python
# Load the tidy ISO3 dataset
df = pd.read_csv('/content/passport-index-tidy-iso3.csv')
df.head()
```

after reading the above mentioned file fetching the required row where: passport=='AGO' means Anagola

```python
# Filter visa requirements where the passport holder is from Angola
angola_visa = df[df['Passport'] == 'AGO']
angola_visa.head()
```

Renamoing the columns

```python
# Rename columns to match Plotly requirements
angola_visa = angola_visa.rename(columns={
    'Destination': 'iso_alpha',    # Needed for Plotly's map
    'Requirement': 'Visa_Status'
})
```

```python
fig = px.choropleth(
    angola_visa,
    locations="iso_alpha",           # ISO3 codes of destination countries
    color="Visa_Status",             # Visa category
    hover_name="iso_alpha",          # Show country code on hover
    color_discrete_map={
        "Visa-Free": "green",
        "Visa on Arrival": "orange",
        "eVisa": "blue",
        "Visa Required": "red",
        "Visa-free": "green",  # case-insensitive catch
        "visa required": "red",
        "e-visa": "blue"
    },
    title="Visa Requirements for Angolan Passport Holders"
)

fig.update_geos(projection_type="natural earth")
fig.update_layout(legend_title_text='Visa Status')
fig.show()
```

Saving The output in html format

```python
fig.write_html("angola_visa_map plotter.html")
from google.colab import files
files.download("angola_visa_map plotter.html")
```
